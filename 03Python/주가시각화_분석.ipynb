```python
import yfinance as yf
import pandas as pd
from sqlalchemy import create_engine, text, inspect
from sqlalchemy.exc import IntegrityError
import pymysql
import os
from dotenv import load_dotenv
load_dotenv()
pymysql.install_as_MySQLdb()
```

```python
# 1. MySQL 데이터베이스 연결 정보
db_host = os.environ.get("DB_HOST")
db_user = os.environ.get("DB_USER")
db_password = os.environ.get("DB_PASSWORD")
db_name =  os.environ.get("DB_name")
table_name = "stock_prices"
engine = create_engine(f"mysql+pymysql://{db_user}:{db_password}@{db_host}/{db_name}")
```

```python
stock_info = pd.read_sql("stock_company_info", con=engine)
stock_info
```

```python
data = pd.read_sql(table_name, con=engine)
data
```

```python
engine.dispose()
```

```python
mged_data = pd.merge(data, stock_info[['종목코드', '회사명', '주식종목']], how='inner', left_on='stock_code', right_on='종목코드')
mged_data
```
