```python
from datetime import datetime, timedelta

yesterday = datetime.today() - timedelta(days=1)
yesterday_date = str(yesterday.date()).replace("-", "")
yesterday_date
```

```python
import ssl
import certifi
import requests
from requests.adapters import HTTPAdapter
from urllib3.poolmanager import PoolManager
import pandas as pd

class TLS12OnlyAdapter(HTTPAdapter):
    def init_poolmanager(self, *args, **kwargs):
        ctx = ssl.create_default_context(cafile=certifi.where())
        # TLS 1.2  (1.3 )
        ctx.minimum_version = ssl.TLSVersion.TLSv1_2
        ctx.maximum_version = ssl.TLSVersion.TLSv1_2
        #    ( )
        ctx.set_ciphers("HIGH:!aNULL:!eNULL:!MD5:!RC4:!DES:!3DES:!DH")
        kwargs["ssl_context"] = ctx
        return super().init_poolmanager(*args, **kwargs)

session = requests.Session()
session.mount("https://", TLS12OnlyAdapter())
session.headers.update({"User-Agent": "Mozilla/5.0"})
url = "https://apis.data.go.kr/1160100/service/GetStockSecuritiesInfoService/getStockPriceInfo"
page_num = 1
total_result = []

while True:
    params = {
        "serviceKey": s_key,  #   %3D%3D   
        "numOfRows": 100,
        "pageNo": page_num,
        "resultType": "json",
        "basDt": yesterday_date,
    }
    r = session.get(url, params=params, timeout=20, verify=certifi.where())
    print(r.status_code)
    # print(r.json())
    response = r.json()
    temp_data = pd.json_normalize(response['response']['body']['items']['item'])
    total_result.append(temp_data)

    total_page = response['response']['body']['totalCount']
    print(f"{page_num}/{total_page} ", end="\r")
    if page_num < total_page:
        page_num += 1
    else:
        break

result_df = pd.concat(total_result)
result_df
```
